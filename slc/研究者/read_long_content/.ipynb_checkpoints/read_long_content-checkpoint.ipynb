```python

def get_time(precision=None):
    """
    获取当前日期和时间

    :param precision: 精确度，可选值为 "seconds" 或 None，默认为 None
    :return: 格式化后的日期和时间

    # 使用示例
    print(get_time())  # 输出格式化后的日期，精确到天
    print(get_time(precision="seconds"))  # 输出格式化后的日期和时间，精确到秒
    """
    current_date = datetime.now()
    if precision == "seconds":
        formatted_date = current_date.strftime("%Y-%m-%d %H:%M:%S")
    else:
        formatted_date = current_date.strftime("%Y-%m-%d")
    return formatted_date
```

```python
from typing import Any, Dict, List, Optional, Sequence, cast
from llama_index.core.schema import BaseNode, TextNode
from llama_index.core.extractors import BaseExtractor
from llama_index.core.extractors import SummaryExtractor
from llama_index.core.async_utils import run_jobs
from tqdm.auto import tqdm

class ReaderExtractor(SummaryExtractor):
    def __init__(self, llm,**kwargs):
        super().__init__(llm=llm,**kwargs,)
    def extract(self, nodes):
        metadata_list = [
            {
                "jingjian": self.llm.complete(f"{node.text}\n\n  ----------- \n 请根据原文内容, 高度浓缩内容为一句话,精简表达").text,
                "person": self.llm.complete(f'\n请以以下格式,从文本中提取人物信息:\n例子:\n    唐三: 一个少年,天才, 家境贫寒\n    小武: 少年天才,拥有完美天赋,和先天满魂力,家境一般\n    周周: 一个胖子,心性不坏,家境未知\n\n文本内容如下:\n-----\n{node.text}\n-----\n').text,
                "thing": self.llm.complete(f'\n请以以下格式,从文本中提取物品信息:\n例子:\n    洗髓丹: 质量一般,价钱合适,用来提升功力\n    天月刀: 神器,开挂版的存在,用来攻击\n    飞盘宝瓶: 逃跑法器,拥有高速移动的能力\n\n文本内容如下:\n-----\n{node.text}\n-----\n').text,
            } for node in tqdm(nodes)
        ]
        return metadata_list
    


    async def _agenerate_node_knowledge(self, node: BaseNode) -> str:
        """Generate a summary for a node."""
        if self.is_text_node_only and not isinstance(node, TextNode):
            return ""

        context_str = node.text
        jingjian =  await self.llm.acomplete(f"{context_str}\n\n  ----------- \n 请根据原文内容, 高度浓缩内容为一句话,精简表达")
        person = await self.llm.acomplete(f'\n请以以下格式,从文本中提取人物信息:\n例子:\n    唐三: 一个少年,天才, 家境贫寒\n    小武: 少年天才,拥有完美天赋,和先天满魂力,家境一般\n    周周: 一个胖子,心性不坏,家境未知\n\n文本内容如下:\n-----\n{context_str}\n-----\n')
        thing = await self.llm.acomplete(f'\n请以以下格式,从文本中提取物品信息:\n例子:\n    洗髓丹: 质量一般,价钱合适,用来提升功力\n    天月刀: 神器,开挂版的存在,用来攻击\n    飞盘宝瓶: 逃跑法器,拥有高速移动的能力\n\n文本内容如下:\n-----\n{context_str}\n-----\n')
        
        result = {
            "jingjian": jingjian.text,
            "person": person.text,
            "thing": thing.text,
        }
        return result


    async def aextract(self, nodes) -> List[Dict]:
        keyword_jobs = []
        for node in nodes:
            keyword_jobs.append(self._agenerate_node_knowledge(node))
        metadata_list = await run_jobs(keyword_jobs, show_progress=self.show_progress, workers=self.num_workers)
        return metadata_list
```

```python
def deal_person_n(person_history):
    # 从 nodes 从获取列表 ,从列表输入 输出处理好的内容
    p = []
    for i,j in person_history:
        for k in j.split('\n'):
            p.append(k)
            
    tt1 = []
    tt3 = []
    for i in p:
        # print(i.split(':'))
        tt1.append(i.split(':')[0])
        tt3.append(i.split(':'))
        
        
    kvcv = {}
    for i in set(tt1):
        xxc = []
        for ix in tt3:
            if ix[0] == i and len(ix)>=2:

                # print(ix[1],'j')
                xxc.append(ix[1])
        kvcv[i] = xxc
        
        
    return {k:v for k,v in kvcv.items() if len(v)>1 and k not in ['周周']}
    # return kvcv
```

```python
import os

def get_folder_paths_with_files(directory):
    folder_paths = []
    for root, dirs, files in os.walk(directory):
        # 确保当前目录下至少有一个文件
        if any(os.path.isfile(os.path.join(root, f)) for f in os.listdir(root)):
            folder_paths.append(root)
    return folder_paths
```
