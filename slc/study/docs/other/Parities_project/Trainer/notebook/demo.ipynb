```python
def get_net(model_path=''):
    input_dim = 8
    condition_dim = 8
    batch = 32

    middle_dim = 50
    hidden_dim = 8
    common_network = nn.Sequential(
        MLP([input_dim + condition_dim,middle_dim])
    )
    mean_network = nn.Sequential(
        MLP([middle_dim,hidden_dim],last_activation=False)
    )
    logvar_network = nn.Sequential(
        MLP([middle_dim,hidden_dim],last_activation=False)
    )
    decode_network = nn.Sequential(
        MLP([hidden_dim+condition_dim,input_dim],last_activation=False),
        nn.Sigmoid(),
    )
    net = CVAE.load_from_checkpoint(model_path,common_network = common_network, 
                                mean_network=mean_network, 
                                logvar_network= logvar_network,
                                decode_network = decode_network,
                                optim=torch.optim.Adam,lr=0.001)
    return net
```
