```python
import RPi.GPIO as GPIO
import time

makerobo_rgbPins = {'Red':32, 'Green':33}  # RGB-LED 管脚定义
makerobo_pirPin = 11    # PIR人体热释电管脚PIN

# 初始化
def makerobo_setup():
    global pir_R, pir_G
    GPIO.setmode(GPIO.BOARD)                 # 将GPIO模式设置为BOARD编号
    GPIO.setwarnings(False)                 # 忽略警告
    GPIO.setup(makerobo_pirPin, GPIO.IN)    # 将makerobo_pirPin设置为输入
    # 将所有LedPin模式设置为输出，初始电平设置为高电平(3.3v)
    for i in makerobo_rgbPins:
        GPIO.setup(makerobo_rgbPins[i], GPIO.OUT, initial=GPIO.HIGH)

    # 设置所有led作为pwm通道和频率到2KHz
    pir_R = GPIO.PWM(makerobo_rgbPins['Red'], 2000)
    pir_G = GPIO.PWM(makerobo_rgbPins['Green'], 2000)

    # 设置所有以值0开始
    pir_R.start(0)
    pir_G.start(0)

# 定义映射值的映射函数。比如从0~255到0~100
def makerobo_MAP(x, in_min, in_max, out_min, out_max):
    return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min

# 定义用于设置颜色的函数
def makerobo_setColor(color):
    # 配置三个led的亮度与输入的颜色值。
    # 区分颜色从“color” 变量修改
    if color == 1:
        pir_R.ChangeDutyCycle(100)
        pir_G.ChangeDutyCycle(0)
    else:
        pir_R.ChangeDutyCycle(0)
        pir_G.ChangeDutyCycle(100)

# 循环函数
def makerobo_loop():
    while True:
        pir_val = GPIO.input(makerobo_pirPin)  # 读取PIR模块管脚电平
        if pir_val==GPIO.HIGH:                 # 如果检测到为高电平
            makerobo_setColor(1)        # 设置RGB-LED灯
        else :
            makerobo_setColor(0)        # 设置RGB—LED灯

# 释放资源
def destroy():
    pir_R.stop()                  # 停止PWM控制
    pir_G.stop()                  # 停止PWM控制
    GPIO.cleanup()                # 释放资源

# 程序入口
if __name__ == '__main__':
    makerobo_setup()   # 初始化
    try:
        makerobo_loop() # 调用循环函数
    except KeyboardInterrupt:  # 当按下Ctrl+C时，将执行destroy()子程序。
        destroy()              # 释放资源
```
