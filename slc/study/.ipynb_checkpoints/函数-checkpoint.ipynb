```python
from sympy import *
x = symbols('x') # 定义符号变量 x
f = Function('f') # 定义函数 f
g = Function('g') # 定义函数 g
expr = f(g(x)) # 构建嵌套函数表达式
print(expr) # 打印表达式
```

```python
expr.diff(x) # 对 x 求导
```

```python
expr = expr.subs({f: Lambda(x, sin(x)), g: Lambda(x, x**2)}) # 用 Lambda 函数替换 f 和 g
print(expr) # 打印表达式
```
