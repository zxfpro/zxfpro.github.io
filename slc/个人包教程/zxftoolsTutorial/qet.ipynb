```python
def get_prompt(system_prompt='你是一个聊天机器人',user_prompt='你好'):
    prompt = """<|begin_of_text|>
<|start_header_id|>system<|end_header_id|>{system_prompt}<|eot_id|>
<|start_header_id|>user<|end_header_id|>{user_prompt}<|eot_id|>
<|start_header_id|>assistant<|end_header_id|>
"""
    return prompt.format(system_prompt='你是一个聊天机器人',user_prompt='你好')
```
