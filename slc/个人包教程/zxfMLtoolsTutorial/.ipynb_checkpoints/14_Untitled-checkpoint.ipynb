```python
import torch
from torch import nn
from enum import Enum

class LrSchedulerType(Enum):
    CosineAnnealingWarmRestarts = 'CosineAnnealingWarmRestarts' #按照token 计数切分
    ReduceLROnPlateau = 'ReduceLROnPlateau' # 语义切分 添加上下文
    CosineAnnealingLR = 'CosineAnnealingLR'

def lr_scheduler(type:LrSchedulerType):

    if type.value == 'CosineAnnealingWarmRestarts':
        scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts
    elif type.value == 'ReduceLROnPlateau':
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau
    elif type.value == 'CosineAnnealingLR':
        scheduler = torch.optim.lr_scheduler.CosineAnnealingLR
    return scheduler
```

```python
import torch
from torch import nn
from enum import Enum

class LossType(Enum):
    CrossEntropyLoss = 'CrossEntropyLoss' #按照token 计数切分
    MSELoss = 'MSELoss' # 语义切分 添加上下文


def get_loss_func(type:LossType):

    if type.value == 'CrossEntropyLoss':
        loss = nn.CrossEntropyLoss()
    elif type.value == 'MSELoss':
        loss = nn.MSELoss()
    return loss
```

```python
import torch
from torch import nn
from enum import Enum

class OptimizerType(Enum):
    SGD = 'SGD' #按照token 计数切分
    Adam = 'Adam' # 语义切分 添加上下文


def get_optimizer_func(type:OptimizerType,model,lr=1e-4):
    if type.value == 'Adam':
        optimizer = torch.optim.Adam(model.parameters(), lr)
    elif type.value == 'SGD':
        optimizer = torch.optim.SGD(model.parameters(), lr)
    return optimizer
```
